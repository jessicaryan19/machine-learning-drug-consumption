### Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(20, 15))
plt.subplots_adjust(hspace=0.8, wspace=0.3)

for idx, drug in enumerate(drug_col, 1):
    plt.subplot(4, 5, idx)
    plt.ylim(0, 2000)
    plt.title(drug)
    plt.xlabel('Usage')
    sns.histplot(data[drug], bins=7, color='darkblue')
    plt.xticks(np.arange(7), ['Never', 'Last Decade', 'Last Year', 'Last Month', 'Last Week', 'Last Day', 'Last Hour'], rotation=45, fontsize=8)
plt.show()
```
