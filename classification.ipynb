### Model
Classification Goal:
1. How likely the person will consume drugs in the future?
2. Which drug does the person use?

```python
import sklearn
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import RobustScaler
```

```python
# import correlation matrix by seaborn
plt.figure(figsize=(20, 15))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset into train and test sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=54)

# Print the shapes of the train and test sets
print("Train set shape:", train_data.shape)
print("Test set shape:", test_data.shape)
```

```python
from sklearn.preprocessing import RobustScaler

# split the data to X and y first
X_train = train_data.drop(drug_col, axis=1)
y_train = train_data[drug_col]

X_test = test_data.drop(drug_col, axis=1)
y_test = test_data[drug_col]

scaler = RobustScaler()

# Fit and transform the data using the scaler
scaled_data = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(y_train)
```

```python
print(scaled_data.shape)
print(y_train.shape)
```

```python
# fit the y_train and scaled_data to the model
model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100)

for drug in drug_col:
    # Train the model
    model.fit(scaled_data, y_train[drug])
    
    # Predict on the train set
    y_train_pred = model.predict(scaled_data)
    y_train_pred = np.round(y_train_pred).astype(int)
    
    # Calculate and print accuracy on train set
    train_accuracy = accuracy_score(y_train[drug], y_train_pred)
    print(f"Drug: {drug}, Train Accuracy: {train_accuracy}")
    
    y_test_pred = model.predict(X_test)
    y_test_pred = np.round(y_test_pred).astype(int)
    
    # Calculate and print accuracy on test set
    test_accuracy = accuracy_score(y_test[drug], y_test_pred)
    print(f"Drug: {drug}, Test Accuracy: {test_accuracy}")
    
```

```python
for drug in drug_col:
    # Predict on the test set
    y_test_pred = model.predict(X_test)
    y_test_pred = np.round(y_test_pred).astype(int)
    
    # Calculate and print accuracy on test set
    test_accuracy = accuracy_score(y_test[drug], y_test_pred)
    print(f"Drug: {drug}, Test Accuracy: {test_accuracy}")
```

```python
# plot the accuracy and loss
plt.figure(figsize=(20, 15))
# plot training and testing accuracy
plt.bar(drug_col, train_accuracy, label='Training Accuracy', color='blue')
plt.show()
```

```python
# plot the accuracy and loss
plt.figure(figsize=(20, 15))
# plot training and testing accuracy
plt.bar(drug_col, test_accuracy, label='Testing Accuracy', color='blue')
plt.show()
```
